```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image, ImageFilter
from skimage import color
from skimage import io
```

```python
image = cv2.imread('C:\\Users\\Nam Le\\sunsetsmall.JPG') # reads the image
image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV) # convert to HSV
# image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
all_gauss_images = [image] + [
    cv2.GaussianBlur(image,(figure_size, figure_size), 0) for figure_size in [3, 7, 15, 35]]

all_median_images = [image] + [
    cv2.medianBlur(image,(figure_size), 0) for figure_size in [3, 7, 15, 35]]

all_mean_images = [image] + [
    cv2.blur(image,(figure_size, figure_size), 0) for figure_size in [3, 7, 15, 35]]

all_bilateral_images = [image] + [
    cv2.bilateralFilter(image, figure_size, 100, 100) for figure_size in [3, 7, 15, 35]]
```

```python
all_blurs = all_gauss_images + all_median_images + all_mean_images + all_bilateral_images

columns = 5
rows = 4
fig = plt.figure(figsize=(4.5, 5))
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(cv2.cvtColor(all_blurs[i-1], cv2.COLOR_HSV2RGB))
    plt.axis("off")
fig.tight_layout()
# plt.show()
plt.savefig('all-sunset-blurs.png', dpi=300)
```

```python
image = cv2.imread('C:\\Users\\Nam Le\\sunset.JPG') # reads the image
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # convert to HSV
# cvtColor( src, src_gray, CV_BGR2GRAY );
new_image = cv2.Laplacian(image, cv2.CV_64F)
```
