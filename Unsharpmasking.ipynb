```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from PIL import ImageFilter
from PIL import Image
import os
```

```python
def lets_unsharp(img_name):
#     file_name = img_name + 'small.jpg'
    image=Image.open(os.path.join('images', img_name +".jpg"))
    new_images = [image, image.filter(ImageFilter.UnsharpMask(radius=5, percent=250)), 
                  image.filter(ImageFilter.UnsharpMask(radius=4, percent=200)), 
                  image.filter(ImageFilter.UnsharpMask(radius=3, percent=150))]
    
    return new_images
```

```python
img_names = ['moon', 'micro', 'blur']
all_unsharp_imgs = []
for img_name in img_names:
    all_unsharp_imgs = all_unsharp_imgs + lets_unsharp(img_name)
```

```python
columns = 4
rows = len(img_names)
fig = plt.figure(figsize=(12.3, 9.5))
titles= ["Original", "Large", "Medium", "Small"]
for i in range(columns*rows):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(all_unsharp_imgs[i])
    plt.axis("off")
    if i<4:
        plt.title(titles[i])
fig.tight_layout()
# plt.show()
plt.savefig(os.path.join('out-images', "all-unsharpmasking-images.pdf"), format='pdf', dpi=300)
```
