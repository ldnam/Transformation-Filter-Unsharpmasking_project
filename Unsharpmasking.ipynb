```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from PIL import ImageFilter
from PIL import Image
import os
```

```python
def lets_histo(img_name, blur_size = 15):
    file_name = img_name + 'small.jpg'
    image=Image.open(os.path.join('images', file_name))
    new_images = [image, image.filter(ImageFilter.UnsharpMask(radius=5, percent=250)), 
                  image.filter(ImageFilter.UnsharpMask(radius=4, percent=200)), 
                  image.filter(ImageFilter.UnsharpMask(radius=3, percent=150))]
    
    return new_images
```

```python
img_names = ['sunset', 'flower', 'house', 'leaf', 'blur']
all_histo_imgs = []
for img_name in img_names:
    all_histo_imgs = all_histo_imgs + lets_histo(img_name)
```

```python
columns = 4
rows = len(img_names)
fig = plt.figure(figsize=(12, 14))
titles= ["Original", "Large", "Medium", "Small"]
for i in range(columns*rows):
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(all_histo_imgs[i])
    plt.axis("off")
    if i<4:
        plt.title(titles[i])
fig.tight_layout()
# plt.show()
plt.savefig(os.path.join('out-images', "all-histogram-equalizations"), dpi=300)
```
