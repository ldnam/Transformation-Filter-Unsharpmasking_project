```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from PIL import ImageFilter
from PIL import Image
```

```python
image=Image.open("C:\\Users\\Nam Le\\flower.jpg")
new_image = image.filter(ImageFilter.UnsharpMask(radius=5, percent=250))
plt.figure(figsize=(11,6))
plt.subplot(121),plt.imshow(image, cmap = 'gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(new_image, cmap = 'gray')
plt.title('Unsharp Filter'), plt.xticks([]), plt.yticks([])
plt.show()
```
